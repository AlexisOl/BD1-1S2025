```python
## instalacion de dependencias
import numpy as np
import matplotlib as mpl
import pandas as pd

## instalacion de dependeica de web scraping
from bs4 import BeautifulSoup
## para tener peticiones
import requests
!pip install selenium
import time
from selenium import webdriver
```

```python
url = "https://es.wikipedia.org/wiki/Departamentos_de_Guatemala"
r = requests.get(url)
# bS
soup = BeautifulSoup(r.content, 'html.parser')

# clase
tabla = soup.find('table', {'class': 'wikitable'})


departamento = []

## CICLO DE ELEMENTOS Y EN BASE A TR y nos saltamos el header
for elementos in tabla.find_all('tr')[1:]:
  valores = elementos.find_all('td')
  for ultimos in valores:
    nombres = ultimos.find_all('th')
    for unicos in nombres:
      departamentos = unicos.find_all('a')
      if (departamentos):
        departamento.append(departamentos)

## iteracin del arreglo
for valoresDepartamento in departamento:
    for a in valoresDepartamento:
        nombre = a.text.strip()
        link = "https://es.wikipedia.org" + a.get('href')
        title = a.get('title')

        print(f"Nombre: {nombre}")
        print(f"Link: {link}")
        print(f"Title: {title}")
        print("-" * 40)

```

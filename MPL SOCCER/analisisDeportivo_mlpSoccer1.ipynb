```python
from mplsoccer import Pitch, FontManager, Sbopen
from matplotlib import rcParams
import matplotlib.pyplot as plt
import pandas as pd
```

```python
messi = pd.read_csv('messibetis.csv')
messi.head()
```

```python
#Tranformacion de una escala mas grande para el analisis
messi['x'] = messi['x']*1.2
messi['y'] = messi['y']*.8
messi['endX'] = messi['endX']*1.2
messi['endY'] = messi['endY']*.8
```

```python
# generacion de campo
pitch = Pitch(pitch_type='statsbomb', pitch_color='#22312b', line_color='#c7d5cc')
fig, ax = pitch.draw(figsize=(16, 11), constrained_layout=True, tight_layout=False)
fig.set_facecolor('#22312b')


pitch.draw(ax=ax)
plt.gca().invert_yaxis()

## ingreso de informacion de pases de messi
for x in range(len(messi['x'])):
  if messi['outcome'][x] == 'Successful':
    plt.plot((messi['x'][x], messi['endX'][x]), (messi['y'][x], messi['endY'][x]), color= 'GREEN')
    plt.scatter(messi['x'][x], messi['y'][x], color= 'GREEN')
  if messi['outcome'][x] == 'Unsuccessful':
    plt.plot((messi['x'][x], messi['endX'][x]), (messi['y'][x], messi['endY'][x]), color= 'RED')
    plt.scatter(messi['endX'][x], messi['endY'][x], color= 'RED')

plt.title("Mapa de Pases de Messi vs Betis", color = "White", size = 20)
```

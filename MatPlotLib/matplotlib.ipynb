```python
## generacion de datos
paises_terremotos = ["Mexico", "Guatemala",
                     "El salvador", "Honduras", "Nicaragua", "Belice", "Costa rica", "Panama", "Estados Unidos"]


valores_terremotos_nivel_1 = np.random.randint(100, 1000, size=len(paises_terremotos))
valores_terremotos_nivel_2 = np.random.randint(100, 1000, size=len(paises_terremotos))+100
valores_terremotos_nivel_3 = np.random.randint(100, 1000, size=len(paises_terremotos))+200
valores_terremotos_nivel_4 = np.random.randint(100, 1000, size=len(paises_terremotos))+300
valores_terremotos_nivel_5 = np.random.randint(100, 1000, size=len(paises_terremotos))+400
```

```python
## imprimir datos
## figure
plt.figure(figsize=(10,8))

plt.plot(paises_terremotos, valores_terremotos_nivel_1, label="Nivel 1")
plt.plot(paises_terremotos, valores_terremotos_nivel_2 , label="Nivel 2")
plt.plot(paises_terremotos, valores_terremotos_nivel_3 , label="Nivel 3")
plt.plot(paises_terremotos, valores_terremotos_nivel_4 , label="Nivel 4")
plt.plot(paises_terremotos, valores_terremotos_nivel_5 , label="Nivel 5")


# leyendas
plt.legend()


# titulo
plt.title("Personas afectadas por terremoto por pais")

# definicion de etiquetas
plt.xlabel("Paises")
plt.ylabel("Personas afectadas")


# ejes de puntuja
plt.yticks(np.arange(0, 1301, 50))


# rendija
plt.grid()

# marcas menores
plt.minorticks_on()

# imprimir
plt.show()
```

# Correlacion de datos

```python
import kagglehub
# Download latest version
path = kagglehub.dataset_download("uciml/pima-indians-diabetes-database")
print("Path to dataset files:", path)
```

```python
#importacion
df = pd.read_csv("/root/.cache/kagglehub/datasets/uciml/pima-indians-diabetes-database/versions/1/diabetes.csv")
```

```python
df.head(10)
```

```python
#valores nulos
df.isnull().sum()
```

```python
df[df['Glucose']== 0]
```

```python
## ingreso de medias
df['Glucose'] = df['Glucose'].replace(0,df['Glucose'].mean())
df.Glucose.value_counts()

```

```python
df[df['BloodPressure'] == 0]['BloodPressure'].value_counts()
df['BloodPressure'] = df['BloodPressure'].replace(0,df['BloodPressure'].mean())
```

```python
df['BMI'] = df['BMI'].replace(0, df['BMI'].median())
```

```python
df['SkinThickness'] = df['SkinThickness'].replace(0, df['SkinThickness'].median())
df['Insulin'] = df['Insulin'].replace(0, df['Insulin'].median())
```

```python
df.describe()
```

```python
plt.figure(figsize=(10,8))
plt.scatter(df["Glucose"], df["BloodPressure"], marker="*")
plt.scatter(df["Insulin"], df["BloodPressure"], marker="+")
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.scatter(df["Glucose"], df["Insulin"], marker="^")
plt.xlabel("Glucose")
plt.ylabel("Insulin")
plt.show()
```

```python
df.corr()
```

```python
plt.scatter(df["BMI"], df["SkinThickness"])
```

```python
plt.matshow(df.corr(), cmap="bwr")
plt.xticks(range(len(df.corr().columns)), df.corr().columns, rotation=90)
plt.yticks(range(len(df.corr().columns)), df.corr().columns)

for i in range(len(df.corr().columns)):
    for j in range(len(df.corr().columns)):
        plt.text(j, i, round(df.corr().iloc[i, j], 2), ha="center", va="center")
plt.show()
```
